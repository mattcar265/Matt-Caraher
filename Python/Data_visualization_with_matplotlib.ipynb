```python
import pandas as pd
import matplotlib.pyplot as plt

mtcars = pd.read_csv('mtcars.csv')

print('Plot Automatic/Manual-based Histogram to compare MPG')

manual_cars = mtcars[mtcars['am'] == 1]['mpg']
automatic_cars = mtcars[mtcars['am'] == 0]['mpg']

plt.figure(figsize=(10, 6))
plt.hist(manual_cars, label='Manual (AM=1)', alpha=0.7)
plt.hist(automatic_cars, label='Automatic (AM=0)', alpha=0.7)

plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')
plt.title('Distribution of MPG by Transmission Type')
plt.legend()

plt.show()
```

```python
print('Use Scatterplot to Plot MPG vs. HP')

mpg = mtcars['mpg']
hp = mtcars['hp']

plt.figure(figsize=(10, 6))

plt.scatter(mpg, hp, color='skyblue')

plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Horsepower (hp)')
plt.title('Scatter Plot of MPG vs Horsepower')

plt.show()
```

```python
print('Create a Scatterplot Matrix for new data')

cols = mtcars[['disp','hp','drat','wt','qsec']]

pd.plotting.scatter_matrix(cols, s = 60)
```

```python
print('Create Boxplots for New Data Consisting of Specified Columns')

plt.figure(figsize=(10, 6))
cols.boxplot()

plt.title('Boxplots of disp, hp, drat, wt, and qsec')
plt.ylabel('Values')
plt.show()
```

```python
print('Use plots to answer which variable has the most impact on MPG')

cols_except_mpg = mtcars.columns.drop('mpg')

for col in cols_except_mpg:
    plt.figure(figsize=(6, 4))
    plt.scatter(mtcars[col], mtcars['mpg'])
    plt.title(f'MPG vs. {col}')
    plt.xlabel(col)
    plt.ylabel('mpg')
    plt.show()
    
```

```python
print('Based on the scatterplots of each variable compared to MPG, I concluded that weight likely has the most impact on MPG. To test this, I will compare it to disp, which I estimated to be the second closest variable in terms of impact on MPG.')

wt_corr = mtcars['mpg'].corr(mtcars['wt'])
disp_corr = mtcars['mpg'].corr(mtcars['disp'])

print('wt correlation to mpg:', wt_corr)
print('disp correlation to mpg:', disp_corr)
```

```python
print('Based on the correlation coefficients, I can conclude that weight does indeed have the highest impact on MPG.')
```
