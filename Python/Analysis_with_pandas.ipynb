```python
print()
print("Analyzing the mpg for cars by gear")

mpg_by_gear = mtcars.groupby('gear')['mpg']

print(mpg_by_gear.describe())

print()
print("Analyzing the mpg for cars by carb")

mpg_by_carb = mtcars.groupby('carb')['mpg']

print(mpg_by_carb.describe())
```

```python
print()
print("Finding out which attribute has the most impact on mpg")

mtcars_numeric_cols = mtcars.select_dtypes(include=[float, int])

corr_matrix = mtcars_numeric_cols.corr()

most_impactful = corr_matrix['mpg']

print(most_impactful)

print()
print("Finding the largest impact:")

print(most_impactful.abs().sort_values())

print("As seen from the listing, the largest impact is wt, which is weight, of the car. This can be seen easier by the sorted list.")
```
